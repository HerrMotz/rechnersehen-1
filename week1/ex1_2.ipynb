## 1. Laden der Bilder
Diesmal sollen mehrere Bilder geladen werden. Zum Auffinden der Bilder soll der oben angegebene Filter `image_filter` eingesetzt werden. Zum Schluss sollten sich alle Bild in einer Liste `images` befinden.

```python
images = [np.asarray(Image.open(x).convert('L')) for x in glob.glob(image_filter)]
```

## 2. Bestimmen von Maximum und Minimum
Definieren Sie eine Funktion `ex1_min_max`, die ein Bild übergeben bekommt. Die Funktion soll den maximalen und den minimalen Grauwert sowie die jeweiligen Koordinaten zurückgeben. Gehen Sie dazu Pixel für Pixel durch das Bild und geben Sie anschließend alle Werte als Tupel zurück.

```python
from typing import Tuple

def ex1_min_max(image: np.ndarray) -> Tuple[float, int, int, float, int, int]:
    """
    Erwartet ein 2D-Graustufenbild (H, W).
    Liefert: (min_wert, min_x, min_y, max_wert, max_x, max_y)
    """
    image_min = float(np.min(image))
    min_index = int(np.argmin(image))
    min_y, min_x = np.unravel_index(min_index, image.shape)

    image_max = float(np.max(image))
    max_index = int(np.argmax(image))
    max_y, max_x = np.unravel_index(max_index, image.shape)

    return image_min, min_x, min_y, image_max, max_x, max_y
```

## 3. Darstellen des Bildes

Wählen Sie eines der Bilder aus der Liste `images` zufällig aus. Verwenden Sie dazu eine Funktion aus dem Paket `numpy`.

```python
# choose random image from images using numpy
# image = np.random.choice(images) # das ging nicht weil
image = choice(images)
```

Funktion zur Ermittlung von Maximum und Minimum verwenden:

```python
image_min, image_min_x, image_min_y, image_max, image_max_x, image_max_y = ex1_min_max(image)
```
